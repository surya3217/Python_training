```python
"""
Challenge 1
    We are going to make a "Shopping List" app. 
    Run the script to start using it.
    Put new things into the list one at a time
    Enter the word DONE - in all CAPS - to QUIT the program
    And once i quit, I want the app to show me everything thats on my list.

Hint 1
    Step 1: Make a list to hold onto our command.
    Step 2: Print out instructions on how to use the app
    Step 3: Ask for new command
    Step 4: Add new command to our list
    Step 5: Be able to quit the app with DONE
    Step 6: print out the list
"""

def add_item(lis):
    '''Add new item in shopping list'''
    lis.append(input('Enter new items: '))

def show(lis):
    '''Display the list items'''
    print("*****Here’s your list*****")
    for item in lis:
        print ( item )
    
def list_help():
    '''Show details of commands'''
    print('''Details of commands: 
             ADD   : add new item in list
             SHOW  : display the current list
             HELP  : give details about the options
             DONE  : quit the app''')

    
# Make a list to hold onto our command.
shopping_list = []

func= ['ADD','SHOW','HELP','DONE']   # list of all commands

while(True):
    
    print('\nList of commands to use Shopping List App:\n',func)
    command = input("Enter the command: ")   # ask for command
    
    if command.upper() in func:
        if command.upper()== 'ADD':
            add_item(shopping_list)
        
        if command.upper()== 'SHOW':
            show(shopping_list)
        
        if command.upper()== 'HELP':
            list_help()
        
        if command.upper()== 'DONE':
            break
    else:
        print('Wrong input! Please try again...')
```
