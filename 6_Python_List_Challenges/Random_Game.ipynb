### 01. Treasure Hunt Game

```python
"""
Problem Statement:
    Create a list of list of integers 10 by 10. In a random position in the array store 
    the number 1. Get the user to enter coordinates where they think the treasure is.If 
    there is a 1 at this position display ‘success’. Repeat Until they find the treasure.
"""
import random

treasure = [[random.randrange(0,2) for x in range(10)] for x in range(10)]    # list comprehension
for i in range(len(treasure)):  
    for j in range(len(treasure[i])):  
        print(treasure[i][j], end="\t") 
    print() 

while(True):
    row, col= input('Enter the coordinates for treasure like (x y): ').split()
    if treasure[int(row)][int(col)]== 1:
        print('Success')
        break
    else:
        print('Loss')
        continue
```

### 02. Code Breaker 

```python
"""
Problem Statement:
    The computer generates a 4 digit code
    The user types in a 4 digit code. Their guess.
    The computer tells them how many digits they guessed correctly
Extension:
    the computer tells them how many digits are guessed correctly in the correct 
    place and how many digits have been guessed correctly but in the wrong place.
    The user gets 12 guesses to either WIN – guess the right code. 
    Or LOSE – run out of guesses.   
"""
from random import randrange

for i in range(13):
    comp = str(randrange(10))+ str(randrange(10))+ str(randrange(10))+ str(randrange(10))
    print('Computer generated 4 digit code:',comp)

    guess= input('Guess the 4 digit code: ')
    corr_pl= 0              # store digits guessed correctly in the correct place
    corr_wr= 0              # store digits guessed correctly but in the wrong place
    
    for ind, digit in enumerate(guess):
        if digit== comp[ind]:
            corr_pl += 1
        elif digit in comp:
            corr_wr += 1

    print('Digits are guessed correctly in the correct place:',corr_pl)
    print('Digits are guessed correctly but in the wrong place:',corr_wr)

    if corr_pl==4:
        print('"WIN" – guess the right code.')
        break
    else:
        print('"LOSE" – run out of guesses.\n')
    
```

### 03. Random Game 2

```python
"""
Problem Statement:
    Write a program for a game where the computer generates a random starting
    number between 20 and 30. The player and the computer can remove 1,2 or 3 
    from the number in turns. Something like this...
    Starting number : 25
    How many do you want to remove? 3
    22 left
    Computer removes 2
    20 left
    The player who has to remove the last value to bring the number
    down to 0 is the loser.
    1 left
    Computer removes 1
    You win!
    Easy option
    Get the computer to choose a number between 1—3 at random
""" 

from random import randrange

num= randrange(20,31)     # generating random number
print('Number:',num)
print('The player and the computer can remove 1,2 or 3 from the number.')
i=0

while(num> 0 and i>=0):
    if i%2==0:               # for even turns player remove value
        while(True):
            try:             # Error handling for non-integer values
                user= int(input('How many do you want to remove?: '))
                if user in range(1,4):
                    break
                else:
                    print('\nInput value is out of range. Please try again...')
            except ValueError:
                print('\nNo valid integer. Please try again...')
                
        num= num- user   
        print('Number:',num)    # displaying the subtracted number
        print()
    else:                       # for odd turns computer remove value  
        comp= randrange(1,4)
        num= num- comp
        print('Computer removed:',comp)
        print('Number:',num)
        print()
    i+=1
    
i-=1
print('Computer won and Player lose.') if i%2==0 else print('Player won and Computer lose.')
```

```python
"""
Problem Statement:
    Write a program for a Higher / Lower guessing game. The computer randomly
    generates a sequence of up to 10 numbers between 1 and 13. The player each 
    after seeing each number in turn has to decide whether the next number is 
    higher or lower. If you can remember Brucie’s ‘Play your cards right’ it’s 
    basically that. If you get 10 guesses right you win the game.
    Starting number : 12
    Higher(H) or lower(L)? L
    Next number 8
    Higher(H) or lower(L)? L
    Next number 11
    You lose
Extension:
    Give the players two lives.Make sure only H or L can be entered 
""" 

from random import randrange

num1= randrange(1,14)       # generating first number between 1 to 13
print(num1)
flag=0
lives= 0                    # variable to store lives

def check(lives):           # function for checking the lives
    if lives==1:
        print('\nThis is your second chance.') 
        return False
    if lives==2:
        print('\nYour lives has finished.')
        return True


for i in range(10):
    guess= input('Enter your guess, Higher(H) or lower(L): ')
    num2= randrange(1,14)        # generating second number between 1 to 13
    print('Next number:',num2)
    
    if (guess=='H' or guess=='h'):  # if guess is 'H' then checking the second number
        if num2> num1:
            num1= num2
            flag+=1             # if guess is correct then up the flag
        else:                   # increment in lives if guess has become wrong
            lives+= 1
            if check(lives):
                break
    elif (guess=='L' or guess=='l'):  # if guess is 'L' then checking the second number
        if num2< num1:
            num1= num2
            flag+=1             # if guess is correct then up the flag 
        else:                   # increment in lives if guess has become wrong
            lives+= 1
            if check(lives):
                break

# checking the flag if flag== i then won else loss the game
print('You guessed all number correctly. You Won.') if flag== i else print('You loss.')
    
```
